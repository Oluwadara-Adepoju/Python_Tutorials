## Matplotlib line

```python
# importing the package

import matplotlib.pyplot as plt#For plotting
%matplotlib inline
```

```python
#Working with a dataset
import pandas as pd
data = pd.read_csv("C:/Users/DELL/Downloads/Projects/water-portability-classification/Train.csv")
data.head()
```

```python
#plotting water turbidity
plt.figure(figsize = (20,15))
plt.plot(data['Turbidity'], ls='dotted')
plt.title('Water Turbidity',size=20)
plt.show()
```

```python
plt.figure(figsize = (20,15))
plt.plot(data['ph'], ls='dashed')
plt.title('Water Ph',size=20)
plt.show()
```

Short format for dot is ':' ,
Dash is '--',
and the long fromat name for linestyle is 'linestyle'

```python
## Line color
plt.figure(figsize = (20,15))
plt.plot(data['Sulfate'], ls = 'dashed', color = 'g')
plt.show()
```

```python
# Line width
# we can change or edit our line width
plt.plot(data['Conductivity'], ls = ':', color = 'hotpink', lw='10.0')
plt.show()
```

```python
#Lets make it bigger
plt.figure(figsize=(20,12))
plt.plot(data['Conductivity'], ls = ':', color = 'c', lw='10.0')
plt.show()
```

## Mulitple line Plots

```python

plt.figure(figsize = (20,15))
plt.plot(data['ph'], ls = 'dashed', color = 'y')
plt.plot(data['Hardness'],  color = 'm')
plt.title('Ph and Hardness',size=20)
plt.show()
```

```python
# We can alos plot multple points
plt.figure(figsize = (20,15))
plt.plot(data['ph'],data['Organic_carbon'],data['Chloramines'],data['Turbidity'])
plt.show()
```

```python
## Creating Labels for Plots

plt.figure(figsize = (20,15))
plt.plot(data['ph'], color = 'y')
plt.plot(data['Hardness'], color = 'm')
plt.title('Ph and Hardness',size=20)
plt.xlabel('Ph', size = 20)
plt.ylabel('Hardness', size = 20)
plt.show()
```

### Creating a title
We can create a title using plt.title()

```python
plt.plot(data['ph'],data['Organic_carbon'],data['Chloramines'],data['Turbidity'])
plt.xlabel('Values')
plt.ylabel('Values')
plt.title('Ph VS Organic Carbon vs Chloramines Vs Turbidity ', size= 20)
plt.show()
```

You can Set font properties for title and labels

```python
import numpy as np
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'New Times Roman','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
# You can als change the title position 
plt.title("Sports Watch Data", loc='right')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.grid(axis = 'y')
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(20,12))
plt.plot(data['Conductivity'], ls = ':', color = 'c', lw='10.0')
plt.grid(axis ='x')
plt.show()
```
