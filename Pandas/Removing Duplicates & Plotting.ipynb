## Removing Duplicates

Another aspect of Data cleaning is removing of duplicate values.  

```python
# importing the necessary packages
import pandas as pd
data = pd.read_csv("C:/Users/DELL/Downloads/Projects/IndabaX/Train.csv")
data.head() 
```

```python
#first we check for duplicates

data.duplicated().sum()/len(data)
```

```python
data.isna().sum()/len(data)*100
```

```python
#Lets say we had duplicated values, we drop duplicated values using the drop_duplicates method
data.drop_duplicates(inplace = True)
```

## Pandas Plotting

We can also use pandas for plotting graphical illustrations

```python
import matplotlib.pyplot as plt#For plotting
%matplotlib inline
data.plot()
plt.show()
```

As seen in the above, there some errors in the data and it mube be cleaned. This are called *outliers*


We can clean the dataset using numpy and outliers

```python

import numpy as np
data['Age'] = np.where(data['Age'].between(-6099,1), 41, data['Age'])#Cleaning the age column
```

```python
#plotting the dataset
data.plot()
plt.show()
```

```python
data.shape
```

```python
data['Age'] = np.where(data['Age'].between(90,320), 41, data['Age'])
data.plot(figsize=(20,12))
plt.title('AutoInland Insurance Data',size=20)
plt.show()
```
