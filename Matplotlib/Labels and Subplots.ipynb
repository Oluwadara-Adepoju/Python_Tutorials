```python
import matplotlib.pyplot as plt#For plotting
%matplotlib inline

import pandas as pd

data = pd.read_csv("C:/Users/DELL/Downloads/Projects/IndabaX/Train.csv")
data.head() 
```

## Creating labels for the Plot

```python
plt.barh(data['ProductName'],data['Age'],color='c',height=0.2)
plt.ylabel('Product Name')
plt.xlabel('Age')
plt.show()
```

## Creating a title Plot

```python
plt.barh(data['ProductName'],data['Age'],color='c',height=0.2)
plt.ylabel('Product Name')
plt.xlabel('Age')
plt.title('Age against Product Name')
plt.show()
```

## Font properties and Font dictionaires

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.barh(data['ProductName'],data['Age'],color='c',height=0.2)
plt.ylabel('Product Name' ,fontdict= font2)
plt.xlabel('Age',fontdict= font2)
plt.title('Age against Product Name',fontdict= font1)
plt.show()
```

```python
plt.barh(data['ProductName'],data['Age'],color='c',height=0.2)
plt.ylabel('Product Name' ,fontdict= font2)
plt.xlabel('Age',fontdict= font2)
plt.title('Age against Product Name',fontdict= font1,loc='right')
plt.show()
```

## Adding Grid Lines

We can use grids to properly indicate our lines where they fall under using plt.grid()

```python
plt.barh(data['ProductName'],data['Age'],height=0.2)
plt.ylabel('Product Name' ,)
plt.xlabel('Age',)
plt.title('Age against Product Name',loc='right')
plt.grid()
plt.show()
```

## Specify your gridline axis
You can indicate your gridline axis

```python
plt.barh(data['ProductName'],data['Age'],height=0.2)
plt.ylabel('Product Name' ,)
plt.xlabel('Age',)
plt.title('Age against Product Name',loc='right')
plt.grid(axis ='y')
plt.show()
```

```python
plt.barh(data['ProductName'],data['Age'],height=0.2)
plt.ylabel('Product Name' ,)
plt.xlabel('Age',)
plt.title('Age against Product Name',loc='right')
plt.grid(axis ='x')
plt.show()
```

```python
plt.barh(data['ProductName'],data['Age'],height=0.2)
plt.ylabel('Product Name' ,)
plt.xlabel('Age',)
plt.title('Age against Product Name',loc='right')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

 ## Subplots

```python
import numpy as np
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title('First')
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('Second')
plt.show()
```
