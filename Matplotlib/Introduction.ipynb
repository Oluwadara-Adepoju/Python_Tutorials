```python
## Format String in Matplotlib
#marker|line|color
ypoints = np.array([2,5, 1,4,8,3 ,10])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python
ypoints = np.array([3,6,7,11,18])
plt.plot(ypoints, 'o-c')
plt.show()
```

```python
ypoints = np.array([3,6,7,11,18])
plt.plot(ypoints, 'o-.m')
plt.show()
```

```python
## Marker Size
plt.plot(ypoints, marker='o', ms =7)
plt.show()
```

```python
# Marker color can also be change
plt.plot(ypoints, marker='o', ms =15, mec='k')# for marker edge color use mec parameter
plt.show()
```

```python
# marker face color
plt.plot(ypoints, marker='o', ms =15, mfc='k')# for marker face color use mfc parameter
plt.show()
```
